### BeautifulSoup
请用 requests 库获取湘潭大学数学与计算科学学院主页信息。
<img src="./xtu-math.png" width = "500" height = "35" div align=center />

```python
headers ={
    'User-Agent':'Mozilla/4.0 (compatible; MSIE S. S; Windows NT)',# 伪装成火狐浏览器
}

url = 'http://math.xtu.edu.cn/'
s = requests.Session()
html = s.get(url)
```

```python
print(html.text)
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(html.text,'lxml')
text_message = soup.find_all('li',class_=['no-sub','nLi on','nLi']) # 选取所有属性class 为‘no-sub'或nLi on或 nLi的标签名为 li 的节点
print(text_message[1])
# 需要的仅为 列表每个元素的直接子节点，标签 a下的文本
# text_message中结构为
#  <li>
#      <a>
#      <div>
#           <ul>
#               <li><a>
```

```python
for i in text_message:
    a=i.a # i 下标签为 a 的节点
    print(a.text)   
```

### JSON + API
实例：请用API遍历方法爬取动态新闻网页。

```python
import json
num = 1  # 爬取的页数
while num<=1:
    url = 'https://feed.sina.com.cn/api/roll/get?pageid=121&lid=1356&num=20&versionNumber=1.2.4&page='+str(num)+'&encode=utf-8'
    news = requests.get(url)
    page = json.loads(news.text)
    data_z = page.get('result').get('data') # 获取result节点下data的数据，记载了新闻的详情
    title=[] # 存储新闻标题
    storge=[]# 存储爬取的新闻文本
    used=[] # 已使用的url
    urls=[] # 未使用的url

    for i in data_z:
        urls.append(i['url'])  #获取新闻url
        title.append(i['title']) # 获取新闻标题
        
    for url_n in urls:
        if url_n not in used: # 去重
            data_url = requests.get(url_n)
            data_url.encoding='utf8' # 因读取的中文出现乱码，进行编码
            used.append(url_n) # 将url加入已爬取队列
            soup = BeautifulSoup(data_url.text,'lxml')
            arctle=soup.find_all('div',class_='article') # 获取属性class=article的div节点，记载了新闻的主体内容
            tem=[]
            for ac in arctle[0].find_all('p'): # 获取div节点下所有p节点，此处存储了新闻的文本
                tem.append(ac.text)
            storge.append('\n'.join(tem)) # 将新闻文本按原格式存储

    num+=1
```
